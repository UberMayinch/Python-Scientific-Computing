## Map:
Map is a mathematical construct that takes an object and returns another object of same type. Usually, the map is iterated multiple times to test for reaching a 'constant' value i.e. steady state

Example: the operation of a map $f$ may written as: <br>
  $x_1=f(x_0)$ <br>
where the map $f$ transform the input $x_0$ to its new value $x_1$.

## Logistical map
Logistic map is a quadratic map; i.e. $f(x)$ is a quadratic ploynomial. 
It shows a particularly interesting phenomena of 'deterministic chaos' i.e. a deterministic map showing apparently random behaviour.  Logistic map has a single parameter, named $\alpha$, and is given by: <br>
$f(x)=\alpha x(1-x)$<br>
When $0\le \alpha \le 4$, the map takes an input $0\le x \le 1$ to give an output in the same range. For a particular value of the parameter $\alpha$, we want to find the behaviour of the map. For such systems, plot of $x_{n}$ vs $x_{n+1}$ is called phase plot and is a important tool for visualising and analysing such systems. 

We will explore this in this tutorial using phase plots.

You can find good detail on Logistic Map at: https://en.wikipedia.org/wiki/Logistic_map

## Solution to Task 1.1

```python
def LogisticMap(alpha, x, num_iters):
    f_x=[x]
    for _ in range(1,num_iters):
        f_x.append(alpha*f_x[_-1]*(1-(f_x[_-1])))
        
    return f_x

alpha=float(input("Enter the value of alpha:"))
x=float(input("Enter the initial value of x"))
num_iters=int(input("Enter the number of iterations"))

print(LogisticMap(alpha, x, num_iters))
plt.plot(LogisticMap(alpha, x, num_iters)), range(num_iters)
plt.show
```

```python
def PhasePlot(x, alpha, num_iters):
    x_line=np.linspace(0,1, 100)
    parabola=[]

    for i in x_line:
        parabola.append(alpha*(i) -alpha*(i**2))

    y_vals=[]
    x_vals=[x]

    for _ in range(num_iters):
        y_vals.append(alpha*(x_vals[_])*(1-x_vals[_]))
        if (_ != num_iters-1):
            x_vals.append(y_vals[_])
    
    
    print(x_vals)
    print(y_vals)
    plt.xlim(0,1)
    plt.ylim(0, 1)
    plt.xlabel("$x_{n}$")
    plt.ylabel("$x_{n+1}$")
    plt.plot(x_line, parabola, alpha=0.5)
    plt.plot(x_line, x_line, alpha=0.5)
    plt.plot(x_vals, y_vals)
    plt.show()


PhasePlot(0.2, 2.9, 100)
```

We see that wherever the phase plot intersects the line $y=x$ corresponding to $x_{n+1}=x_{n}$ we have a solution. Also we have drawn the parabola $y=\alpha x - \alpha x^{2}$ for reference and we see that the phase plot will never go outside this parabola. The maximum value of our function for the phase plot will be $\frac{\alpha}{4}$

### Strategy to solve this task: to find the steady-state or cycle for a given $\alpha$

1. Find roots of $f^{(n)}-x=0$, where $f^{(n)}$ is the $n^{th}$ iteration of the map; i.e. $f^{(3)}(x)=f(f(f(x)))$
2. Check the stability of the roots. 
1. If all roots are unstable, then increase n by 1 and goto step 1
1. If even one root is stable, then we have found the steady-state/cycle.

Alternatively: 
$ \frac{d}{dx} f^{(n)}(x) = \frac{df(x_{n-1})}{dx}  \frac{d}{dx} f^{(n-1)}(x)  = \left[\alpha (1-2 x_{n-1})\right] \frac{d}{dx} f^{(n-1)}(x_{n-1}) $, which is a recursive formula for the derivative!

$$\begin{eqnarray}
\text{ Denote } x_n &= f^{(n)}(x) \\
\frac{d}{dx} f^{(n)}(x) & = \frac{d}{dx} x_n = \frac{d}{dx} f(x_{(n-1)}) =
                     \left[\alpha (1-2 x_{(n-1)})\right] \frac{d}{dx} x_{(n-1)} \\
 \text{ with } x_0 = x \implies \frac{d}{dx} x_0 = 1
\end{eqnarray}
$$

$$ \frac{d}{dx} f^{(n)}(x)  = \frac{d}{dx} x_n = \frac{d}{dx} f(x_{(n-1)}) = f^\prime(x_{(n-1)}) \frac{d}{dx} x_{(n-1)} \text{ where }f^\prime(x)=\alpha (1-2x) \text{ and }\frac{d}{dx}x_0 = 1$$

```python

def checkStability(alpha, num_iters):
    x = sp.symbols('x', real=True)
    a = sp.symbols('a', real=True)

    expr=LogisticExpressions[num_iters-1].subs(a, alpha)
    solutions = sp.solve(expr - x, x)

    derivative = Derivatives[num_iters-1].subs(a,alpha)
    for sol in solutions:
        derivative_value = derivative.subs(x, sol)
        if abs(derivative_value) > 1:
            continue
        else:
            return True
        
    return False
```

```python

def iterativeLogMap(alpha, num_iters):
    x = sp.symbols('x', real=True)
    expr = x
    for _ in range(num_iters):
        expr = alpha * expr * (1 - expr)

    return expr


def findIteratedSoln(alpha, num):
    x = sp.symbols('x', real=True)
    if ((checkStability(alpha, num) == False) and (num <= 5)):
        print(f"The map is unstable with {num} composition(s)")
        return findIteratedSoln(alpha, num + 1)
    else:
        solutions=sp.solve(iterativeLogMap(alpha, num) - x, x)
        return solutions, num
    
    
result, num=findIteratedSoln(3.5, 1)
print("The stable states are", result)
print(f"There is a {2 ** (num-1)} cycle for this value of alpha")

```

```python
def logistic_function(alpha, x):
    return alpha*(1-x)*x

def BifurcationGraph(start, end, num_steps, iterations, initial):
    fig, axs = plt.subplots()
    fig.set_size_inches(24, 15)
    for alpha in np.linspace(start, end, num_steps):
        y_given_x = [initial]
        for i in range(1, iterations):
            y_given_x.append(logistic_function(alpha, y_given_x[i-1]))
        plt.plot([alpha]*300, y_given_x[200:], 'g.', markersize = 0.05)
    plt.xlabel("$Alpha$")
    plt.ylabel("Steady States")
    plt.title('Bifurcation Diagram')
    plt.show()

BifurcationGraph(2.5, 4, 10000, 500, 0.3)
```

Here we run 500 iterations at each value of x and the density of the points plotted at one given value is proportional to the number of steady states. We see that initially all the steady states cluster are in 1 or 2 branches but with successive bifurcation the density keeps on decreasing corresponding to the number of cycles increasing. 

The bifurcation diagram depicts the behavior of the logistic map at different alpha values, revealing its stable states and long-term dynamics. It provides a visual representation of how the system transitions from stability to chaos as the parameter is varied.It is the graph we obtain when we plot the steady states as a function of x. 

Key observations from the bifurcation diagram of the logistic map can be summarized as follows:

1. Stable Regimes: At specific alpha values, the logistic map settles into a steady state where it remains fixed or oscillates between a few values. These stable regimes are represented by horizontal lines or regions in the bifurcation diagram. This is in the region $\alpha < 3$

2. Period Doubling: Increasing alpha leads to a period-doubling bifurcation in the logistic map. This means that stable regimes split into two stable values, which further split into four, then eight, and so on, resulting in a cascading pattern of periodic behavior. The bifurcation diagram exhibits this phenomenon as a branching structure. The periods start increasing at a faster and faster rate due to splitting of cycles. The number of stable points keep increasing and we need to plot iteratively more composed versions of the function to plot the stable states of the map. This is typically in the region $3 < \alpha < 3.7$

3. Chaotic Regimes: Beyond a critical alpha value, the logistic map enters a chaotic region. In this region, the system demonstrates sensitive dependence on initial conditions, where even slight changes in the starting state result in drastically different long-term behavior. There is no stochasticity involved however, so this is called *deterministic chaos*. Chaotic regimes in the bifurcation diagram appear as intricate and dense patterns, displaying seemingly random or irregular distributions of points. This is the region $\alpha > 3.7$.
